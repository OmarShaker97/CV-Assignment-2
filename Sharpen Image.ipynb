```python
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open("cameraman.tif").convert("L")
 
sharpened_img = img.copy()

averaging_value = 0
 
# CALCULATE FOR ALL PIXELS THAT ARE ON THE LEFT    
for i in range (0,img.size[0]):
    for j in range (0,img.size[1]):
        try:
            top_pixel = img.getpixel((i, j-1))
        except:
            top_pixel = 0
            
        try:
            bottom_pixel = img.getpixel((i,j+1))
        except:
            bottom_pixel = 0
            
        try:
            right_pixel = img.getpixel((i+1, j))
        except:
            right_pixel = 0
            
        try:
            left_pixel = img.getpixel((i-1,j))
        except:
            left_pixel = 0
            
        try:
            top_left_pixel = img.getpixel((i-1, j-1))
        except:
            top_left_pixel = 0
       
        try:
            top_right_pixel = img.getpixel((i+1, j-1))    
        except:
            top_right_pixel = 0
            
        
        try:
            bottom_right_pixel = img.getpixel((i+1, j+1))
        except:
            bottom_right_pixel = 0
      
        try:
            bottom_left_pixel = img.getpixel((i-1,j+1))
        except:
            bottom_left_pixel = 0
            
        averaging_value = (img.getpixel((i,j))*8 + (top_pixel*-1) + (bottom_pixel*-1) + (left_pixel*-1) + (right_pixel*-1) + (top_left_pixel*-1) + (top_right_pixel*-1) + (bottom_right_pixel*-1) + (bottom_left_pixel*-1))//9
        
        sharpened_img.putpixel((i,j),img.getpixel((i,j)) + averaging_value)
        
        
       
plt.imshow(sharpened_img, cmap='gray')
sharpened_img.save('Sharpened.jpg')
```
